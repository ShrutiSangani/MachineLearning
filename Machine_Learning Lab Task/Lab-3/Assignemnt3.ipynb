Group data using custom function: Let's say you want to group your data using custom function. Here the requirement is to create three groups

1. Days when temperature was between 80 and 90
2. Days when it was between 50 and 60
3. Days when it was anything else

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("weather_by_cities.csv")
df
```

```python
new_df=df
```

Making new Column in dataset

```python
def categorize_temperature(temp):
    if 80<=temp<=90:
        return "80-90"
    elif 50<=temp<=60:
        return "50-60"
    else:
        return "other"
```

```python
new_df['Temperature Group'] = new_df['temperature'].apply(categorize_temperature) 
```

```python
new_df
```

```python
groupe_data=new_df.groupby('Temperature Group',as_index=False)
```

```python
for group, group_data in groupe_data:
    print("Group : ",group)
    print(group_data)
    print()
```

Categorized in groupby function

```python
groupe_data2=df.groupby([df['temperature'].between(80,90),df['temperature'].between(50,60)])
```

```python
for group,group_data in groupe_data2:
    print("Category : (80-89, 50,60) : ",group)
    print(group_data)
    print()
```

# Filling missing values

```python
import pandas as pd
```

```python
df=pd.read_csv("weather_data.csv")
print(df)
```

```python
df=df.set_index('day')
```

```python
new_df=df
```

```python
#replace null value with 0
new_df=df.fillna(0)
```

```python
new_df
```

```python
new_df=df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No event'
})
new_df
```

```python
new_df=df.fillna(method='bfill')
```

```python
new_df
```

```python
new_df=df.fillna(method='ffill')
```

```python
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)
#there we give limit to fill only one forward fill
```

```python
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1,axis=1)
```

```python
new_df
```

```python
new_df=df.interpolate()
```

```python
new_df
```

# dropping row

```python
new_df=df.dropna()
```

```python
new_df
```

```python
new_df=df.dropna(how='all')
#delete row which contains all null value
new_df
```

```python
new_df=df.dropna(thresh=1)
#drop not null 1
```

```python
new_df
```

```python
new_df=df.dropna(thresh=2)
new_df
```

```python
df.dropna(thresh=0.3*len(df.columns))
#it will drop those rows which rows have 30% data only
```

```python
df.dropna(thresh=7,axis=1)
```

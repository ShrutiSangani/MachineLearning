**emoji used as marker & using image**

```python
x = np.array([0,2,4,6])
y = np.array([3,7,5,12])
plt.plot(x,y, marker='$\U0001F601$', markersize = 20)
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
image=plt.imread('emoji.png')
image_box = OffsetImage(image, zoom=0.05)
```

```python
fig, ax = plt.subplots()
x, y = np.atleast_1d(x, y)
 
# for each value of (x,y), we create
# an annotation
for x0, y0 in zip(x, y):
    ab = AnnotationBbox(image_box, (x0, y0), frameon=False)
    ax.add_artist(ab)
 
ax.plot(x,y, c='purple')
plt.show()
```

**overlap barchart**

```python
bars = ['00', '01', '02', '03', '04', '05']
part1 = [12, 9, 5, 7, 8, 2]
part2 = [4, 16, 8, 20, 12, 9]
```

```python
x_axis = np.arange(len(bars))
plt.bar(x_axis +0.2, part1, width=0.4, label = 'part1', color='purple')
plt.bar(x_axis -0.2, part2, width=0.4, label = 'part2', color='orange')   
plt.xticks(x_axis, bars)
plt.legend()
plt.show()
```

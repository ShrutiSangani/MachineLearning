```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
# Python Program to illustrate Linear Plotting
# year contains the x-axis values and e-india & e-bangladesh are the y-axis values for plotting

 

year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]

 

# Setting figure size
fig = plt.figure(figsize = (8, 6)) 
# plotting of x-axis(year) and y-axis(power consumption) with different colored labels of two countries


#(x-axis,y-axis,colour,label,marker,markersize)

plt.plot(year, e_india, color ='orange',label ='India',marker ='*', markersize = 12) #marker ='o', markersize = 12, 

 

plt.plot(year, e_bangladesh, color ='g',label ='Bangladesh',linestyle ='dashed', linewidth = 2) # linestyle ='dashed', linewidth = 2,

 

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.grid()
# naming of x-axis and y-axis
plt.xlabel('Years',fontdict = font1)
plt.ylabel('Power consumption in kWh',fontdict = font1)

 

# naming the title of the plot
plt.title('Electricity consumption per capita of India and Bangladesh',fontdict = font2)

 

pop_a = mpatches.Patch(color='#5DC83F', label='India')
pop_b = mpatches.Patch(color='#CE5D45', label='Bangladesh')
plt.legend(handles=[pop_a,pop_b],frameon=False,labelspacing = 2,ncol=2)
# plt.legend(handles=[pop_a,pop_b],frameon=True,labelspacing = 20,ncol=1)

 
#for default

# plt.legend()

#for legend
#The attribute Loc in legend() is used to specify the location of the legend.Default value of loc is loc=”best” (upper left). 
#The strings ‘upper left’, ‘upper right’, ‘lower left’, ‘lower right’ place the legend at the corresponding corner of the axes/figure.

 

#The attribute bbox_to_anchor=(x, y) of legend() function is used to specify the coordinates of the legend, 
#and the attribute ncol represents the number of columns that the legend has.It’s default value is 1.

#where to locate legend inside of chart

# plt.legend(["Coutry-1", "Country-2"], loc ="lower right")

#where to locate legend outside chart
# plt.legend(bbox_to_anchor =(0.75, 1.25), ncol = 1)


# plt.fill_between(e_india, e_bangladesh, year, color='green', alpha=0.5)
plt.show()
```
